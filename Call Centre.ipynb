```python
#How many calls were within SLA and above SLA?
# Define the SLA threshold (e.g., 300 seconds)
sla_threshold = 300

# Calculate calls within SLA and above SLA
calls_within_sla_and_above = Df[(Df['call duration in minutes'] <= sla_threshold) & (Df['call duration in minutes'] > 0)]

# Get the count
num_calls_within_sla_and_above = len(calls_within_sla_and_above)

print("Number of calls within SLA and above SLA:", num_calls_within_sla_and_above)
```
