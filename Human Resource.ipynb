```python
#What is the total number of employees based on race?
hr.groupby('race')['id'].count().reset_index().sort_values (by= 'id', ascending = False)
```

```python
#Which department has the highest number of staff?
hr.groupby('department')['id'].count().reset_index().sort_values (by='id', ascending = False)
```

```python
#In the engineering department, do we have more male or females?
engineering_df = hr[hr['department'] == 'Engineering']

# Count the number of males and females in the engineering department
male_count = (engineering_df['gender'] == 'Male').sum()
female_count = (engineering_df['gender'] == 'Female').sum()

print("Number of males in Engineering:", male_count)
print("Number of females in Engineering:", female_count)
```

```python
#Most staff are from what location?
hr.groupby('location')['id'].count()
```

```python
#How many staff have their employment termed?s
termination_count = hr['termdate'].isna().sum() + (hr['termdate'] == 'UTC').sum()

print("Number of staff members with employment termed:", termination_count)



```

```python
#How many employees are due for promotion? (To qualify for promotion, you must have worked for 7 years or above)
(hr['hire year'] <= 2016).sum()
```

```python
#How many employees are due for retirement? (To qualify, you must be 50 years and above

current_year = dt.datetime.now().year

# Calculate the age of each employee
df= current_year - hr['Birth Year']

# Count the number of employees due for retirement (age 50 and above)
(df >= 50).sum()
```
